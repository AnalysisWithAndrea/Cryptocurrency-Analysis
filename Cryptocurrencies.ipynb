```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
doge = pd.read_csv(r"C:\Users\andre\Downloads\DOGE-USD.csv")
btc = pd.read_csv(r"C:\Users\andre\Downloads\BTC-USD (1).csv")
eth = pd.read_csv(r"C:\Users\andre\Downloads\ETH-USD (1).csv")
ada = pd.read_csv(r"C:\Users\andre\Downloads\ADA-USD.csv")

doge.info()
```

```python
doge[['Year','Month','Day']] = doge['Date'].str.split('-', expand=True)
btc[['Year','Month','Day']] = btc['Date'].str.split('-', expand=True)
eth[['Year','Month','Day']] = eth['Date'].str.split('-', expand=True)
ada[['Year','Month','Day']] = ada['Date'].str.split('-', expand=True)
```

```python
plt.figure(figsize=(15,7))
plt.subplots_adjust(top=1.25, bottom=1.2)
plt.subplot(2,2,1)
doge['Close'].plot()
plt.subplot(2,2,2)
btc['Close'].plot()
plt.subplot(2,2,3)
eth['Close'].plot()
plt.subplot(2,2,4)
ada['Close'].plot()
plt.tight_layout()
```

```python
plt.figure(figsize=(15,7))
plt.subplots_adjust(top=1.25, bottom=1.2)
plt.subplot(2,2,1)
doge['Volume'].plot()
plt.subplot(2,2,2)
btc['Volume'].plot()
plt.subplot(2,2,3)
eth['Volume'].plot()
plt.subplot(2,2,4)
ada['Volume'].plot()
plt.tight_layout()
```

```python
ma_day = [10, 20, 50]
crypto_list = [doge,btc,eth,ada]
for ma in ma_day:
    for crypto in crypto_list:
        column_name = f"MA for {ma} days"
        crypto[column_name] = crypto['Adj Close'].rolling(ma).mean()
doge.head(30)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
fig.set_figheight(8)
fig.set_figwidth(15)

doge[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(ax=axes[0,0])
axes[0,0].set_title('Doge')

btc[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(ax=axes[0,1])
axes[0,0].set_title('Bitcoin')

eth[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(ax=axes[1,0])
axes[0,0].set_title('Etherium')

ada[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(ax=axes[1,1])
axes[0,0].set_title('Cardano')

fig.tight_layout()
```

```python
for crypto in crypto_list:
    crypto['Daily Return'] = crypto['Adj Close'].pct_change()
fig, axes = plt.subplots(nrows=2, ncols=2)
fig.set_figheight(8)
fig.set_figwidth(15)

doge['Daily Return'].plot(ax=axes[0,0], legend=True, linestyle='--', marker='o')
axes[0,0].set_title('Doge')

btc['Daily Return'].plot(ax=axes[0,1], legend=True, linestyle='--', marker='o')
axes[0,1].set_title('Bitcoin')

eth['Daily Return'].plot(ax=axes[1,0], legend=True, linestyle='--', marker='o')
axes[1,0].set_title('Etherium')

ada['Daily Return'].plot(ax=axes[1,1], legend=True, linestyle='--', marker='o')
axes[1,1].set_title('Cardano')

fig.tight_layout()
```

```python
plt.figure(figsize=(12, 7))
names = ['Doge','Bitcoin','Etherium','Cardano']
for i, crypto in enumerate(crypto_list, 1):
    plt.subplot(2, 2, i)
    crypto['Daily Return'].hist(bins=50)
    plt.ylabel('Daily Return')
    plt.title(f'{names[i-1]}')
    
plt.tight_layout()
```

```python
close_df = pd.DataFrame()
close_df['Doge'] = doge['Close']
close_df['Bitcoin'] = btc['Close']
close_df['Etherium'] = eth['Close']
close_df['Cardano'] = ada['Close']
close_df
```

```python
rets = close_df.pct_change()
rets
```

```python
sns.jointplot(x='Doge', y='Bitcoin', data= rets, kind='scatter')
```

```python
sns.pairplot(rets, kind='reg')
```

```python
sns.heatmap(rets.corr(), annot=True, cmap='ocean')
```

```python
rets = rets.dropna()
area = np.pi * 20
plt.figure(figsize=(10, 7))
plt.scatter(rets.mean(), rets.std(), s=area)
plt.xlabel('Expected return')
plt.ylabel('Risk')
for label, x, y in zip(rets.columns, rets.mean(), rets.std()):
    plt.annotate(label, xy=(x,y), xytext=(50,50), textcoords='offset points', ha='right', va='bottom', arrowprops=dict(arrowstyle='-', color='blue', connectionstyle='arc3,rad=-0.3'))
```
