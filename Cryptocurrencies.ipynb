```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#loading in the data
doge = pd.read_csv(r"C:\Users\andre\Downloads\DOGE-USD.csv")
btc = pd.read_csv(r"C:\Users\andre\Downloads\BTC-USD (1).csv")
eth = pd.read_csv(r"C:\Users\andre\Downloads\ETH-USD (1).csv")
ada = pd.read_csv(r"C:\Users\andre\Downloads\ADA-USD.csv")

doge.info()
```

```python
print(doge.shape)
doge.head()
```

```python
#formatting date column
doge[['Year','Month','Day']] = doge['Date'].str.split('-', expand=True)
btc[['Year','Month','Day']] = btc['Date'].str.split('-', expand=True)
eth[['Year','Month','Day']] = eth['Date'].str.split('-', expand=True)
ada[['Year','Month','Day']] = ada['Date'].str.split('-', expand=True)
```

```python
doge.describe()
```

```python
# determining date of highest price for Doge
doge.iloc[doge['High'].idxmax()]
```

```python
# determining date of lowest price for Doge
doge.iloc[doge['Low'].idxmin()]
```

```python
#setting date column as the index od the dataframe
doge.set_index('Date', inplace=True)
btc.set_index('Date', inplace=True)
ada.set_index('Date', inplace=True)
eth.set_index('Date', inplace=True)
```

```python
#plotting Doge Adj close traded over time
doge.truncate(before='2020-03-25', after='2022-03-25')['Adj Close'].plot(legend=True,figsize=(12,5))
```

```python
#plotting Closing value over time
plt.figure(figsize=(17,9))
plt.subplots_adjust(top=1.25, bottom=1.2)
plt.subplot(2,2,1)
doge['Close'].plot(title='Doge')
plt.subplot(2,2,2)
btc['Close'].plot(title='Bitcoin')
plt.subplot(2,2,3)
eth['Close'].plot(title='Ethereum')
plt.subplot(2,2,4)
ada['Close'].plot(title='Cardano')
plt.tight_layout()
```

```python
#Plotting the total volume traded over time
plt.figure(figsize=(17,9))
plt.subplots_adjust(top=1.25, bottom=1.2)
plt.subplot(2,2,1)
doge['Volume'].plot(title='Doge',legend=True)
plt.subplot(2,2,2)
btc['Volume'].plot(title='Bitcoin',legend=True)
plt.subplot(2,2,3)
eth['Volume'].plot(title='Ethereum',legend=True)
plt.subplot(2,2,4)
ada['Volume'].plot(title='Cardano',legend=True)
plt.tight_layout()
```

```python
#Calculating moving average over a 10, 20, and 50 day period
ma_day = [10, 20, 50]
crypto_list = [doge,btc,eth,ada]
for ma in ma_day:
    for crypto in crypto_list:
        column_name = f"MA for {ma} days"
        crypto[column_name] = crypto['Adj Close'].rolling(ma).mean()
doge.head(30)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
fig.set_figheight(9)
fig.set_figwidth(17)

doge[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(ax=axes[0,0])
axes[0,0].set_title('Doge')

btc[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(ax=axes[0,1])
axes[0,1].set_title('Bitcoin')

eth[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(ax=axes[1,0])
axes[1,0].set_title('Ethereum')

ada[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(ax=axes[1,1])
axes[1,1].set_title('Cardano')

fig.tight_layout()
```

```python
#calculating and plotting the daily return 
for crypto in crypto_list:
    crypto['Daily Return'] = crypto['Adj Close'].pct_change()
    
fig, axes = plt.subplots(nrows=2, ncols=2)
fig.set_figheight(8)
fig.set_figwidth(15)

doge.truncate(before='2020-03-25', after='2022-03-25')['Daily Return'].plot(ax=axes[0,0], legend=True, linestyle='--', marker='o')
axes[0,0].set_title('Doge')

btc.truncate(before='2020-03-25', after='2022-03-25')['Daily Return'].plot(ax=axes[0,1], legend=True, linestyle='--', marker='o')
axes[0,1].set_title('Bitcoin')

eth.truncate(before='2020-03-25', after='2022-03-25')['Daily Return'].plot(ax=axes[1,0], legend=True, linestyle='--', marker='o')
axes[1,0].set_title('Ethereum')

ada.truncate(before='2020-03-25', after='2022-03-25')['Daily Return'].plot(ax=axes[1,1], legend=True, linestyle='--', marker='o')
axes[1,1].set_title('Cardano')

fig.tight_layout()
```

```python
#histogram of daily return
plt.figure(figsize=(15, 8))
names = ['Doge','Bitcoin','Ethereum','Cardano']
for i, crypto in enumerate(crypto_list, 1):
    plt.subplot(2, 2, i)
    sns.distplot(crypto['Daily Return'].dropna(), bins=43, color='blue').set(title=names[i-1],ylabel= 'Density')
    
    
plt.tight_layout()
```

```python
#creating a dataframe with closing values for all 4 cryptos
close_df = pd.DataFrame()
close_df['Doge'] = doge['Close']
close_df['Bitcoin'] = btc['Close']
close_df['Ethereum'] = eth['Close']
close_df['Cardano'] = ada['Close']
close_df
```

```python
#calculating closing percentage change
rets = close_df.pct_change()
rets
```

```python
sns.pairplot(rets, kind='reg')
```

```python
plt.figure(figsize=(13,8))
#used a Pearson correlation because we are dealing with an interval scale
sns.heatmap(rets.corr(method='pearson'), annot=True, cmap='ocean')
```

```python
plt.figure(figsize=(13,8))
sns.heatmap(close_df.corr(method='pearson'), annot=True, cmap='ocean')
```

```python
#calculating and plotting risk of investing in each coin
rets = rets.dropna()
area = np.pi * 20
plt.figure(figsize=(15, 7))
#using the mean of the coin's daily returns as the expected return
plt.scatter(rets.mean(), rets.std(), s=area)
plt.xlabel('Expected return')
plt.ylabel('Risk')
for label, x, y in zip(rets.columns, rets.mean(), rets.std()):
    plt.annotate(label, xy=(x,y), xytext=(50,50), textcoords='offset points', ha='right', va='bottom', arrowprops=dict(arrowstyle='->', color='red', connectionstyle='arc3,rad=-0.5'))
```
